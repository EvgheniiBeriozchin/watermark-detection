Import packages

```python
import random
import cv2 as cv
import os
import pandas as pd
```

```python
path_images = '/content/drive/MyDrive/briquet_engraving'
images = os.listdir(path_images)
Nr_images = len(images)
```

Create random images

```python
Random_img = random.sample(range(Nr_images), 1000)
```

```python
for i in range(1000):
  path = os.path.join(path_images, images[i])
  img = cv.imread(path,cv.IMREAD_GRAYSCALE)
  filename = '/content/TestImages/test_images_' + str(i+1) + '.png'
  cv.imwrite(filename, img)
```
